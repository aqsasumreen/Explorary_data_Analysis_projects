```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# The UnicodeDecodeError you're encountering suggests that the CSV file contains characters that are not encoded in UTF-8. To resolve this, you can try specifying a different encoding when reading the CSV file. Common alternative encodings include 'ISO-8859-1' (also known as 'latin1').
df  = pd.read_csv('Diwali_Sales_Data.csv' , encoding='unicode-escape') # to avoid encoding errors
# in case of any error write r' in the start of file name 
df.head(5)
```

```python
df.info()
```

```python
# Missing  values analysis
df.isnull().sum()
```

```python
df.drop(['unnamed1', 'Status'],axis= 1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Amount'].fillna(df['Amount'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x = 'Gender', data = df) # first data analysis
plt.title('count of order priorities')
plt.show()
```

```python
# Bivariate
plt.figure(figsize=(4,4))
sns.countplot(x = 'Gender', data = df, hue='Age Group') # first data analysis
plt.title('count of order priorities')
plt.show()
```
